## How Sentence Transformers models work


```python
from sentence_transformers import SentenceTransformer, models

## Step 1: use an existing language model
word_embedding_model = models.Transformer('m3hrdadfi/bert-fa-base-uncased-wikinli-mean-tokens')

## Step 2: use a pool function over the token embeddings
pooling_model = models.Pooling(word_embedding_model.get_word_embedding_dimension())

## Join steps 1 and 2 using the modules argument
model = SentenceTransformer(modules=[word_embedding_model, pooling_model])
```

```python
corpus = [
' دستگاه روشن شود ',
'دستگاه خاموش شود',
'سرعت دستگاه کم شود',
'سرعت دستگاه زیاد شود',
           'سرعت تغییر کند'
# 'چراغها روشن شود ',
# 'چراغها خاموش شود'
]

# Query sentences:
queries = [
'ازت میخوام دستگاه روشن نشه',
 'ازت میخوام دستگاه روشن شه',

            'rate تولید سازه متریال هندلینگ به ۶ بشه',
            'میزان تولید متریال افزایش پیدا کنه',
'دستگاه پی آی دی خاموش نکن',
            'سرعت بع مقدار ۵ بره',
            'دستگاه شروع به فعالیت کنه',
            'دستگاه از فعالیت بایسته',
            'دستگاه متوقف بشه',
            'ساز متريال همدلي و روشن',
            ' MATERIAL AND رو روشن نكن'
]
top_k = 2
```

```python
from sentence_transformers import models, SentenceTransformer, util
import torch
corpus_embeddings = modelB.encode(corpus, convert_to_tensor=True, show_progress_bar=True)
```

```python
for query in queries:
    query_embedding = modelB.encode(query, convert_to_tensor=True, show_progress_bar=True)
    cos_scores = util.pytorch_cos_sim(query_embedding, corpus_embeddings)[0]
    cos_scores = cos_scores.cpu()

    #We use torch.topk to find the highest 5 scores
    top_results = torch.topk(cos_scores, k=top_k)

    print(f'Query: {query}')

    for score, idx in zip(top_results[0], top_results[1]):
        print(f'{corpus[idx]} --- (Score: {score:.4f})')

    print('- - ' * 10)
```
